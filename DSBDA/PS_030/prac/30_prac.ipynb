```python
soup=bs(req.text,"html.parser")
soup
```

```python
reviews=soup.findAll("div",{"class":"t-ZTKy"})
all_rev=[]
for review in reviews:
    rev_text=review.get_text().replace("READ MORE","").strip()
    # strip removes the useless white spaces from the start and end of sentence
    all_rev.append(rev_text)
    print(rev_text+"\n\n")
```
